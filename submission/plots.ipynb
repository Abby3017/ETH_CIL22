```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ['Logistic Regression', 'Bidirectional LSTM', 'ELECTRA', 'DistilBERT', 'Combined GELU']
small_dataset = [81.2, 81.475, 86.525, 87.825, 87.795]
full_dataset = [83.2, 81.792, 89.016, 88.916, 89.16]

fs='x-large'

x = np.arange(len(labels))  # the label locations
width = 0.42  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, small_dataset, width, label='Small Dataset', color='#D6604D')
rects2 = ax.bar(x + width/2, full_dataset, width, label='Full Dataset', color='#4393C3')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Accuracy', fontsize='x-large')
#ax.set_title('Accuracy Scores')
ax.set_xticks(x, labels)
ax.set_yticks(range(80, 91, 2))
ax.set_ylim([80, 90])
ax.legend(fontsize=fs, frameon=False)

ax.bar_label(rects1, padding=3, fontsize=fs)
ax.bar_label(rects2, padding=3, fontsize=fs)

fig.tight_layout()

plt.setp(ax.get_xticklabels(), rotation=20, ha='right', fontsize='x-large')

plt.rcParams["figure.figsize"] = (12,7)

plt.savefig('accuracy_gb_models.png', bbox_inches="tight")
plt.show()
```

```python
datasets = ['Small Dataset', 'Full Dataset']
models = ['Logistic Regression', 'Bidirectional LSTM', 'ELECTRA', 'DistilBERT', 'Combined GELU']
colors = ['#003f5c', '#58508d', '#bc5090', '#ff6361', '#ffa600']

data = np.array(([81.2, 83.2], [81.472, 81.792], [86.525, 89.016], [87.825, 88.916], [87.795, 89.16]))
fw = np.array

fs='x-large'

x = np.arange(len(datasets))  # the label locations
width = 0.18  # the width of the bars

fig, ax = plt.subplots()

for i in range(len(models)):
    rects = ax.bar(x + (i-2)*width, data[i], width, align='center', label=models[i], color=colors[i])
    labels = ax.bar_label(rects, padding=3, fontsize=fs)
    
    if i==3:
        labels[0].set(fontweight='bold')
    elif i==4:
        labels[1].set(fontweight='bold')

    
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Accuracy', fontsize=fs)
#ax.set_title('Accuracy Scores')
ax.set_xticks(x, datasets)
ax.set_yticks(range(80, 91, 2))
ax.set_ylim([80, 90])
ax.legend(fontsize=fs, frameon=False)

fig.tight_layout()

plt.setp(ax.get_xticklabels(), fontsize=fs)

plt.rcParams["figure.figsize"] = (12,7)

plt.savefig('accuracy_gb_datasets.png', bbox_inches="tight")
plt.show()
```
